### total_load

```python
# total load

df_load_2020 = pd.read_csv('/Users/emircanince/Desktop/data/total_load/total_load_2020.csv')
df_load_2021 = pd.read_csv('/Users/emircanince/Desktop/data/total_load/total_load_2021.csv')
df_load_2022 = pd.read_csv('/Users/emircanince/Desktop/data/total_load/total_load_2022.csv')
df_load_2023 = pd.read_csv('/Users/emircanince/Desktop/data/total_load/total_load_2023.csv')
df_load_2024 = pd.read_csv('/Users/emircanince/Desktop/data/total_load/total_load_2024.csv')
```

```python
df_load = pd.concat([df_load_2020, 
                     df_load_2021, 
                     df_load_2022, 
                     df_load_2023, 
                     df_load_2024], axis=0, ignore_index=True)
```

```python
df_load.head(10)
```

```python
# preprocess

df_load['Date'] = df_load['MTU (CET/CEST)'].str.split(' - ').str[0]
df_load = df_load[['Date', 'Day-ahead Total Load Forecast (MW)']]

print(df_load.head(10))
```

```python
df_load['Date'] = pd.to_datetime(df_load['Date'], dayfirst=True)
df_load['Hour'] = df_load['Date'].dt.floor('H')

df_load = df_load.groupby('Hour', as_index=False)['Day-ahead Total Load Forecast (MW)'].sum()
df_load.rename(columns={'Hour': 'Date'}, inplace=True)

print(df_load.head())
```

```python
df_total_load = df_solar_forecast.iloc[:, :1].copy()
df_total_load.head()
```

```python
df_total_load['Date'] = pd.to_datetime(
    df_total_load['Date'],
    dayfirst=True,
    infer_datetime_format=True,
    errors='coerce'
)

df_load['Date'] = pd.to_datetime(df_load['Date'])

df_total_load = df_total_load.merge(
    df_load[['Date', 'Day-ahead Total Load Forecast (MW)']],
    on='Date',
    how='left'
)

df_total_load.rename(columns={df_total_load.columns[1]: 'total_load'}, inplace=True)

df_total_load['total_load'] = df_total_load['total_load'].round(2)

print(df_total_load.head())
```

```python
df_total_load['Date'] = pd.to_datetime(df_total_load['Date'])
print(df_total_load.shape)
print(df_total_load.dtypes)
```

### aggregate

```python
dfs = [
    df_solar_forecast,
    df_wind_off_forecast,
    df_wind_on_forecast,
    df_total_load,
    df_capacity,
    df_electricity_price,
    df_daylight_hours,
    df_co2_price,
    df_gas_price
]

for df in dfs:
    df.drop_duplicates(subset='Date', inplace=True)
```

```python
df_thesis = df_solar_forecast.iloc[:, :1].copy()

for df in dfs:
    df_thesis = df_thesis.merge(df, on='Date', how='left')

print(df_thesis.shape)
display(df_thesis.head())
```

```python
df_thesis['co2_price'] = pd.to_numeric(df_thesis['co2_price'], errors='coerce')
```

```python
df_thesis['wind_forecast'] = (df_thesis['wind_on_forecast'] + df_thesis['wind_off_forecast']).round(2)
df_thesis['wind_capacity'] = (df_thesis['wind_on_capacity'] + df_thesis['wind_off_capacity']).round(2)
df_thesis['solar_penetration'] = (df_thesis['solar_forecast'] / df_thesis['total_load'] * 100).round(2)
df_thesis['wind_penetration'] = (df_thesis['wind_forecast'] / df_thesis['total_load'] * 100).round(2)
```

```python
df_thesis['Hour'] = df_thesis['Date'].dt.hour
df_thesis['Day'] = df_thesis['Date'].dt.day
df_thesis['Month'] = df_thesis['Date'].dt.month
df_thesis['Year'] = df_thesis['Date'].dt.year
```
